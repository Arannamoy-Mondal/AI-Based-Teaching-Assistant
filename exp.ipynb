```python
from langchain_ollama import OllamaEmbeddings,ChatOllama
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import Chroma
from langchain_classic.chains import RetrievalQA
from langchain_text_splitters import RecursiveCharacterTextSplitter
import os
```

```python
embedding_model=OllamaEmbeddings(model="nomic-embed-text:latest")
```

```python
chat_model=ChatOllama(model="nemotron-cascade-2:latest")
```

```python
files=os.listdir(path="pdf")
```

```python
files
```

```python
pdf_loader=PyPDFLoader(f"pdf/{files[0]}")
data=pdf_loader.load()
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=100)
pages=text_splitter.split_documents(data)
```

```python
vector_db=Chroma.from_documents(documents=pages,embedding=embedding_model)
```

```python
qa_chain=RetrievalQA.from_chain_type(llm=chat_model,retriever=vector_db.as_retriever())
```

```python
response=qa_chain.invoke("Explain this paper in bangla.")
```
